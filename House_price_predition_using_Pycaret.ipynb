```
import pandas as pd
```

```
train_df = pd.read_csv('/content/drive/My Drive/House price prediction - April/train.csv')
```

```
test_df = pd.read_csv('/content/drive/My Drive/House price prediction - April/test.csv')
```

```
from pycaret.regression import *

reg1 = setup(data = train_df, target = 'SalePrice')
```

```
linear_model = create_model('lr')
```

```
final_model = finalize_model(linear_model)
```

```
predictions = predict_model(final_model, data = test_df)
```
